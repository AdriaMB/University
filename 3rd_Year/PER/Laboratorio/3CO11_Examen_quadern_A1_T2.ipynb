**Ejercicio:** $\;$ Empleando el clasificador de Análisis Discriminante Cuadrático (QDA) visto en las sesiones de laboratorio, se alcanza una tasa de acierto del 95.93% en la tarea de MNIST. Con el objetivo de **intentar** mejorar esta tasa de acierto, completa el siguiente cuaderno para realizar un experimento con QDA en el que, variando la dimensionalidad de los datos mediante PCA, se haga una exploración de los valores del hiperparámetro de regularización (`reg_param`). Completa la variable `dni` con las tres últimas cifras de tu DNI sin letra para que sea usada como semilla de inicialización en la exploración de hiperparámetros.

```python
import warnings; warnings.filterwarnings("ignore")
import numpy as np
import datasets
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, ShuffleSplit, RandomizedSearchCV
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from scipy.stats import loguniform
```

```python
dni=308 # Completar
```

```python
ds = datasets.load_dataset("ylecun/mnist").with_format("numpy")
X_train = ds['train'][:]['image'].astype(np.float32).reshape(-1, 28*28) / 255.
y_train = ds['train'][:]['label'].astype(np.uint8)
X_test = ds['test'][:]['image'].astype(np.float32).reshape(-1, 28*28) / 255.
y_test = ds['test'][:]['label'].astype(np.uint8)
```

```python
max_K = np.min(X_train.shape)
pca = PCA(n_components=max_K).fit(X_train) # Completar

X_train = pca.transform(X_train) # Completar
X_test = pca.transform(X_test) # Completar

Ks = np.array([50, 100, 200, 300, 400, 500, max_K]) # Completar

clf = QuadraticDiscriminantAnalysis()
G = {'reg_param': loguniform(1e-9, 1e-1)} # Completar
splitter = ShuffleSplit(n_splits=1, test_size=0.1, random_state=dni)
RS = RandomizedSearchCV(clf, G, n_iter=3, scoring='accuracy', n_jobs=-1, cv=splitter, pre_dispatch=32, random_state=dni)
for _, k in enumerate(reversed(np.sort(Ks))):
    Z_train = X_train[:,:k] # Completar
    Z_test = X_test[:,:k] # Completar

    acc = RS.fit(Z_train, y_train).score(Z_test, y_test) # Completar
    print(f'Precisión: {acc:.2%} con {RS.best_params_} y K = {k}')
```

A la vista de los resultados, qué combinacion de valores para `K` y `reg_param` elegirías? Por qué?

 ya que ofrecen la mejor precisión de todas las combinaciones (95.92%) al con una reducción de dimensión notable (784 -> 100)
