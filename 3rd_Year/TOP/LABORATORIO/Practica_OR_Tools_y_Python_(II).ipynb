# **1. Un problema de distribución de vehículos**

Una compañía fabrica automóviles en 3 plantas y los envía a 4 regiones del país. Las plantas pueden suministrar las cantidades que se indican en la siguiente tabla (Capacidad). Las demandas de los clientes por región se incluyen en la última fila de la tabla, así como los costes de enviar un automóvil desde cada planta a cada región.
La compañía desea conocer cuál es el plan de distribución a coste mínimo con el que satisfacer las demandas de las 4 regiones sin sobrepasar las capacidades de las plantas fabricantes.

 **a)**	Plantea un **modelo de programación lineal** que permita a la compañía determinar el plan de distribución con el **coste mínimo**, cumpliendo la capacidad de las plantas y las demandas regionales. Obtén la solución óptima con OR-TOOLS (GLOP) + PYTHON.

```python
#Instalación de OR Tools
!pip install ortools

#Importar el wrapper de OR Tools para trabajar con solvers de programación lineal
from ortools.linear_solver import pywraplp

#Creación de la instancia del solver
solver = pywraplp.Solver.CreateSolver('GLOP')  # GLOP es un solver LP. 

# Datos del problema
coste = [
    [131, 218, 266, 120],  # Planta 1
    [250, 116, 263, 278],  # Planta 2
    [178, 132, 122, 180]   # Planta 3
]

capacidad = [450, 500, 300]      # Capacidades de las plantas
demanda = [450, 200, 300, 300]   # Demandas de las regiones

# Variables de decisión: x[i][j]
x = []
for i in range(3):
    x.append([])
    for j in range(4):
        var = solver.NumVar(0, solver.infinity(), f'x_{i+1}_{j+1}')
        x[i].append(var)

# Restricciones de CAPACIDAD de plantas
for i in range(3):
    solver.Add(sum(x[i][j] for j in range(4)) <= capacidad[i])

# Restricciones de DEMANDA de regiones
for j in range(4):
    solver.Add(sum(x[i][j] for i in range(3)) >= demanda[j])

# FUNCIÓN OBJETIVO: minimizar coste total
objective = solver.Objective()
for i in range(3):
    for j in range(4):
        objective.SetCoefficient(x[i][j], coste[i][j])
objective.SetMinimization()

# Resolver
tipo_resultado=solver.Solve()

# Informar sobre el tipo de resultado
print("-----------------------------------------------")

if tipo_resultado == solver.OPTIMAL :
  print("SE HA ENCONTRADO LA SOLUCIÓN ÓPTIMA")
  print("-----------------------------------------------")
  for i in range(3):
    for j in range(4):
       print(f'Cantidad a enviar desde la Planta {i+1} a la Región {j+1}: {x[i][j].solution_value()}')
  print("-----------------------------------------------")
  print("Valor óptimo de la función objetivo es", solver.Objective().Value())
  print("-----------------------------------------------")

if tipo_resultado == solver.ABNORMAL :
  print("Se ha producido un error mientras se ejecutaba el solver")

if tipo_resultado == solver.INFEASIBLE :
  print("El problema no tiene solución posible")

if tipo_resultado == solver.FEASIBLE :
  print('Coste total =', solver.Objective().Value())
  for i in range(3):
    for j in range(4):
       print(f'Cantidad a enviar desde la Planta {i+1} a la Región {j+1}: {x[i][j].solution_value()}')
  print("-----------------------------------------------")
  print("Valor óptimo de la función objetivo es", solver.Objective().Value())
  print("-----------------------------------------------")

if tipo_resultado == solver.NOT_SOLVED :
  print("No se ha podido encontrar ninguna solución en el tiempo proporcionado")

if tipo_resultado == solver.UNBOUNDED :
    print("Solución no acotada por las restricciones")
```

Por cada automóvil demandado en cada región y no enviado, se incurre en una penalización de 150, 20, 10 y 80 para las regiones, 1, 2, 3 y 4 respectivamente. Modifica el modelo del apartado c) de modo que se asegure la **máxima distribución de automóviles** con el **coste mínimo**. ¿En qué regiones y en qué cantidad no se podrá satisfacer la demanda?

```python
#Importar el wrapper de OR Tools para trabajar con solvers de programación lineal
from ortools.linear_solver import pywraplp

#Creación de la instancia del solver
solver = pywraplp.Solver.CreateSolver('GLOP')  # GLOP es un solver LP. 

# Datos del problema
coste = [
    [131, 218, 266, 120],  # Planta 1
    [250, 116, 263, 278],  # Planta 2
    [178, 132, 122, 180]   # Planta 3
]

capacidad = [450, 500, 300]      # Capacidades de las plantas
demanda = [450, 200, 300, 350]   # Demandas de las regiones
penalizacion = [150, 20, 10, 80]  # Penalizacino por no cumplir demanda en cada region

# Variables de decisión: x[i][j], y[j]
x = []  # Coches fabricados por planta i para region j
y = []  # Demanda no suplida de la region j
for i in range(3):  # por cada planta
    x.append([])
    for j in range(4):  #  por cada region
        var = solver.NumVar(0, solver.infinity(), f'x_{i+1}_{j+1}')
        x[i].append(var)

for j in range(4):
  var = solver.NumVar(0, solver.infinity(), f'y_{j+1}')
  y.append(var)


# Restricciones de CAPACIDAD de plantas
for i in range(3):
    solver.Add(sum(x[i][j] for j in range(4)) == capacidad[i])

# Restricciones de DEMANDA de regiones
for j in range(4):
    solver.Add(sum(x[i][j] for i in range(3)) + y[j] == demanda[j])

# FUNCIÓN OBJETIVO: minimizar coste total
objective = solver.Objective()
for i in range(3):
    for j in range(4):
        objective.SetCoefficient(x[i][j], coste[i][j])
for j in range(4):
  objective.SetCoefficient(y[j], penalizacion[j])
objective.SetMinimization()

# Resolver
tipo_resultado=solver.Solve()

# Informar sobre el tipo de resultado
print("-----------------------------------------------")

if tipo_resultado == solver.OPTIMAL :
  print("SE HA ENCONTRADO LA SOLUCIÓN ÓPTIMA")
  print("-----------------------------------------------")
  for i in range(3):
    for j in range(4):
       print(f'Cantidad a enviar desde la Planta {i+1} a la Región {j+1}: {x[i][j].solution_value()}')
  print("-----------------------------------------------")
  for j in range(4):
    print(f'La demanda no cubierta de la planta {j+1} es: {y[j].solution_value()}')
  print("-----------------------------------------------")  
  print("Valor óptimo de la función objetivo es", solver.Objective().Value())
  print("-----------------------------------------------")

if tipo_resultado == solver.ABNORMAL :
  print("Se ha producido un error mientras se ejecutaba el solver")

if tipo_resultado == solver.INFEASIBLE :
  print("El problema no tiene solución posible")

if tipo_resultado == solver.FEASIBLE :
  print('Coste total =', solver.Objective().Value())
  for i in range(3):
    for j in range(4):
       print(f'Cantidad a enviar desde la Planta {i+1} a la Región {j+1}: {x[i][j].solution_value()}')
  print("-----------------------------------------------")
  print("Valor óptimo de la función objetivo es", solver.Objective().Value())
  print("-----------------------------------------------")

if tipo_resultado == solver.NOT_SOLVED :
  print("No se ha podido encontrar ninguna solución en el tiempo proporcionado")

if tipo_resultado == solver.UNBOUNDED :
    print("Solución no acotada por las restricciones")
```
