```python
import string

def is_pangrama(phrase: str) -> bool:
    return set(phrase.lower()).intersection(set(string.ascii_lowercase)) == set(string.ascii_lowercase)

is_pangrama("hello"),\
is_pangrama("The quick brown fox jumps over the lazy dog")



```

```python
def perf_pangrama(phrase: str) -> bool:
    return is_pangrama(phrase) and len(phrase.lower().replace(" ", "")) == len(string.ascii_lowercase)

perf_pangrama("The quick brown fox jumps over the lazy dog"),\
perf_pangrama("abcdefghij klmnopq    rstuvw xyz")
```

```python
def funcion_grep(document: str, word: str) -> list:
    res = []
    try:
        f = open(document, "r", encoding="utf-8")
        content = f.read()
        content = content.split("\n")
        for line in content:
            if word in line:
                res.append(line)
    except FileNotFoundError as e:
        print("File ", document," not found")
    finally:
        return res


print(funcion_grep("prueba.txt", "la")),\
print(funcion_grep("prueddba.txt", "la"))
```

```python
def char_freq_base(word: str) -> dict:
    return { i:word.lower().count(i) for i in list(string.ascii_lowercase)}


def char_freq_improved(word: str) -> dict:
    res = {}
    for i in list(string.ascii_lowercase):
        if i in word:
            res.update({i:word.lower().count(i)})
    return res

def char_freq(word:str) -> dict:
    return {i:word.lower().count(i) for i in list(string.ascii_lowercase) if i in word}

char_freq_base("abbabcbdbabdbdbabababcbcbab"),\
char_freq_improved("abbabcbdbabdbdbabababcbcbabz"),\
char_freq("abbabcbdbabdbdbabababcbcbab")
```

```python
def is_balanced(word:str)->str:
    counter = 0
    for i in word:
        if i == "[":
            counter+=1
        elif i == "]":
            if counter == 0:
                return "NOT OK"
            else:
                counter-=1
        else:
            continue
    
    return "OK" if counter==0 else "NOT OK"

is_balanced("[]"),\
is_balanced("]["),\
is_balanced("[hola][]]][[]]")
```
